```python
import csv
```

```python
if __name__ == "__main__":
    
    with open("final_mal_data.csv", "w", newline ="") as writefile:
        writer = csv.writer(writefile)
        reader = open("mal_data.csv","r")
        it = 0
        
        for dataset in reader:
            if it == 0:
                it += 1
                continue
            
            li = []
            
            for i in dataset:
                if i == '1':
                    li.append(i)
                elif i == '0':
                    li.append(-1)
                    
            print(li)
            writer.writerow(li)

        reader.close()
```
